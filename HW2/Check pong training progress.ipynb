```python
import numpy as np
```

```python

reward_data_file = 'reward_trend124.npy'

reward_data = np.load(reward_data_file)


from matplotlib import pyplot as plt
plt.plot(reward_data, color='blue')
plt.plot(np.convolve(reward_data, np.ones((100,))/100, mode='valid'),color = 'red')
plt.show()
```
